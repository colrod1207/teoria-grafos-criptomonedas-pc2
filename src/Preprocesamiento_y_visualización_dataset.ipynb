# **Preprocesamiento de los Datos**

```python
#Conversión de tipos de datos
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')
df['market_cap'] = pd.to_numeric(df['market_cap'], errors='coerce')
df['total_volume'] = pd.to_numeric(df['total_volume'], errors='coerce')
```

```python
# Verificar si existen valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

```python
#----------------TRATAMIENTO DE DATOS FALTANTES-------------------------

# Eliminar filas sin market_cap o current_price
df = df.dropna(subset=['market_cap', 'current_price'])

# Imputar valores faltantes con técnicas como imputación de la mediana o reemplazar con valores predeterminados
df['total_volume'] = df['total_volume'].fillna(df['total_volume'].median())
df['circulating_supply'] = df['circulating_supply'].fillna(df['circulating_supply'].median())
df['max_supply'] = df['max_supply'].fillna(-1)
df['rank'] = df['rank'].fillna(-1)

# Imputar cambios porcentuales con 0 (sin cambio)
df['price_change_percentage_24h'] = df['price_change_percentage_24h'].fillna(0)
df['market_cap_change_percentage_24h'] = df['market_cap_change_percentage_24h'].fillna(0)

# Imputar precios altos/bajos con current_price (si falta)
df['high_24h'] = df['high_24h'].fillna(df['current_price'])
df['low_24h'] = df['low_24h'].fillna(df['current_price'])

# Imputar ATH y ATL con medianas
df['ath'] = df['ath'].fillna(df['ath'].median())
df['atl'] = df['atl'].fillna(df['atl'].median())

# Imputar fecha con la más reciente conocida
df['last_updated'] = df['last_updated'].fillna(df['last_updated'].max())
```

```python
# Verificar si existen valores faltantes nuevamente
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

```python
#Eliminar datos duplicados según el id
df = df.drop_duplicates(subset=['id'])
```

```python
# Visualización de outliers con boxplots antes de eliminarlos
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
plt.figure(figsize=(8, 4))
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
plt.show()
```

```python
# Guardar el DataFrame limpio en un nuevo archivo CSV
df.to_csv("criptos_limpios.csv", index=False)
print("Dataset limpio guardado como 'criptos_limpios.csv'")
```

# **Visualizaciones del dataset**

```python
# Ver cantidad de proyectos por categoría
print("\nCantidad de proyectos por categoría:")
display(df['category'].value_counts())
```

```python
# Gráfico de barras - Cantidad de proyectos por categoría
plt.figure(figsize=(8, 5))
df['category'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Cantidad de proyectos por categoría")
plt.xlabel("Categoría")
plt.ylabel("Número de proyectos")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión: market cap vs. volumen (opcional)
plt.figure(figsize=(8, 5))
plt.scatter(df['market_cap'], df['total_volume'], alpha=0.5)
plt.title("Market Cap vs. Volumen Total")
plt.xlabel("Market Cap")
plt.ylabel("Volumen Total")
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.tight_layout()
plt.show()
```
